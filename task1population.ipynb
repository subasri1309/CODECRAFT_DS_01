```python
import pandas as pd
import matplotlib.pyplot as plt
```

copying the path of the file that has been added in the colab

```python
data= pd.read_csv("/content/data set.csv", skiprows=4, on_bad_lines="skip")
```

understanding the data

```python
data.shape
```

```python
data.columns = data.columns.str.strip()
```

```python
print(data.columns)
```

```python
population_2023 = data["2023"].dropna()
```

showing the columns in the form of bar graph

```python
plt.figure(figsize=(10, 6))
plt.hist(population_2023, bins=30, edgecolor="black", alpha=0.7)
plt.xlabel("Population in 2023")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population Across Countries (2023)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
```
